question 1.
        
        Explain the difference between Relative, Absolute and Mixed Cell Referencing.
  
  Answere 1 --

     #   (a) -  Relative cell referencing

     - Change automatically when copied to another cell.

     - Example: If =A1 + B1 in cell C1 is copied to C2, it becomes =A2 + B2.

     - Use case: Great for repeating calculations across rows or columns.


     #   (b) -  Absolute cell referencing

     - Remain fixed no matter where the formula is copied.

     - Written with $ sign → $A$1.

     - Example: If =A1 * $B$1 is copied down, only A1 changes, $B$1 stays constant.

     - Use case: Useful for fixed values like tax rates, conversion factors, constants.


     #   (c)  -  Mixed cell referencing

    - Combination of relative and absolute.

    - $A1 → column A fixed, row changes.

    - A$1 → row 1 fixed, column changes.

    - Use case: Helpful in tables or multiplying across both rows and columns.
